### The Dataset

We will be working with the [NBA Stats (1947-present)](https://www.kaggle.com/datasets/sumitrodatta/nba-aba-baa-stats/data) Dataset, created by Sumitro Datta. The dataset contains 22 separate csv files, containing a total of 32 MB of data. It contains player, team, and award data from the past 78 NBA seasons.

We are currently unsure exactly what we plan on predicting, due to feasibility concerns. Our leading ideas are:
* Predicting +/- rating, offensive rating, defensive rating, or position based on player simple stats and team.
* Predicting all-star, all-NBA, and all-defensive teams based on previous players, team, and current player stats.
* Predicting draft pick position (lottery, first round, second round, undrafted) based on team, age, and stats.


### Implementation/technical merit

Anticipated challenges: The NBA and basketball is notoriously unpredicted, with some players being undrafted and winning a DPOY or second-overall picks being out of the league in a year. Due to the unpredictability, it might be difficult to predict and classify certain attributes. Especially with rule changes, injuries, and roster changes not being cataloged in the dataset.

Due to the dataset being 78 years long, we will need to pare down the data significantly. We plan on removing at least every season before the year 2000, and possibly removing more. It is important to remove previous years, because play styles, stats, and competition has greatly increased over the lifetime of the NBA.


### Potential impact of the results

If our classification and predictions are accurate enough, this can become a useful tool for people interested in fantasy sports or sports betting. It will also be interesting if we can predict future results of awards or team records for the current season and see it's predictive accuracy.
